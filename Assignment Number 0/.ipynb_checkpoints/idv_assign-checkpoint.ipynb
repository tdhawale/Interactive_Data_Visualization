```python
import numpy as np
import matplotlib.pyplot as plt
from PIL import Image
%matplotlib inline
by_lst = []
with open('colorado_elev.vit', 'rb') as f:
    while True:
        data = f.read(1)
        if not data:
            break
        #img = np.append(img,data)
        by_lst.append(int.from_bytes(data, byteorder='big'))

#img = Image.fromarray(data, 'RGB')
#mg.show()
```
